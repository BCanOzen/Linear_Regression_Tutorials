# Polynomial Regression
----------------------

# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
dataset = pd.read_csv("/kaggle/input/position-salaries/Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

Lets draw the Graph of target(Salary).

```python
plt.scatter(X,y,color = "blue")
plt.title("Salary Graph")
plt.xlabel("level")
plt.ylabel("Salary")
```

From graph as you can see the targets are not linear.

# Training the Linear Regression model on the whole dataset

We calculate the linear regression to see the difference.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

# Training the Polynomial Regression model on the whole dataset

PolynomialFeatures from sklearn.preprocessing is calculating polynomial regression. PolynomialFeatures(degree = 4) is choosing degree of the n from polynomial linear regression equation.

Y = bo + b1.X1^1 + b2.X1^2 + .... + bnX1^n

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

# Visualising the Linear Regression results

As you can see from graph linear regression line is not we can use for model.

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title("POLYNOMIAL REGRESSION")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

# Visualising the Polynomial Regression results

```python
plt.scatter(X,y,color = "red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)) , color = "blue")
plt.title("POLYNOMIAL REGRESSION")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

# Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting a new result with Linear Regression

Orijinal value of 6.5 position level is like 160.000 but in linear regression says 330.000.

```python
lin_reg.predict([[6.5]])
```

# Predicting a new result with Polynomial Regression

As you can see polynomial regression result is more close to target.

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
